```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
file_path = 'pose_landmarks.csv'  # Replace with your CSV file name
data = pd.read_csv(file_path)

# Choose a joint to plot
joint_index = 20  # Replace with the desired joint index (0-32)

# Get the column names for x, y, and z of the selected joint
x_column = f'{joint_index}_x'
y_column = f'{joint_index}_y'
z_column = f'{joint_index}_z'

# Apply a rolling mean to smooth the data
window_size = 5  # Set the size of the moving average window
data[x_column] = data[x_column].rolling(window=window_size).mean()
data[y_column] = data[y_column].rolling(window=window_size).mean()
data[z_column] = data[z_column].rolling(window=window_size).mean()

# Plot the smoothed joint movement
plt.figure(figsize=(10, 6))
plt.plot(data['Frame'], data[x_column], label=f'Joint {joint_index} (X)', color='red')
plt.plot(data['Frame'], data[y_column], label=f'Joint {joint_index} (Y)', color='green')
plt.plot(data['Frame'], data[z_column], label=f'Joint {joint_index} (Z)', color='blue')
plt.xlabel('Frame')
plt.ylabel('Coordinates (Smoothed)')
plt.title(f'Smoothed Movement of Joint {joint_index} (X, Y, Z)')
plt.legend()
plt.grid()
plt.show()
```
